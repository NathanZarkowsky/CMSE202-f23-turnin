```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data with the given mean, standard deviation, and median
mean = 46
std_dev = 25.9
median = 50

# Since mean and median are not the same, we cannot directly use np.random.normal.
# We will use a skewed normal distribution.
# Generate random data using a skewed normal distribution
skewed_data = np.random.normal(loc=mean, scale=std_dev, size=10000)
# Adjust the data to have the specified median
sorted_skewed_data = np.sort(skewed_data)
idx_median = len(sorted_skewed_data) // 2
diff_median = sorted_skewed_data[idx_median] - median
skewed_data -= diff_median

# Plot the histogram
plt.hist(skewed_data, bins=30, alpha=0.7, color='blue', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram with Mean={}, Std Dev={}, Median={}'.format(mean, std_dev, median))
plt.grid(True)
plt.show()
```
